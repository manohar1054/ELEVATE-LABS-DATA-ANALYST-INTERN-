```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
server = r'DESKTOP-39CALVC\SQLEXPRESS'
database = 'sales_data_sample'
driver = 'ODBC Driver 17 for SQL Server'
```

```python
connection_string = f"mssql+pyodbc://@{server}/{database}?driver={driver}&trusted_connection=yes"
engine = create_engine(connection_string)
```

```python
df = pd.read_sql("SELECT TOP 10 * FROM sales_data_sample", engine)
df.head()
```

```python
Show all tables in the database
```

```python
df = pd.read_sql("""
    SELECT TABLE_NAME 
    FROM INFORMATION_SCHEMA.TABLES 
    WHERE TABLE_TYPE = 'BASE TABLE'
""", engine)
df
```

```python
 Show column names and data types for a specific table
```

```python
df = pd.read_sql("""
    SELECT COLUMN_NAME, DATA_TYPE 
    FROM INFORMATION_SCHEMA.COLUMNS 
    WHERE TABLE_NAME = 'your_table_name'
""", engine)
df
```

```python
pd.read_sql("SELECT * FROM INFORMATION_SCHEMA.TABLES", engine)
```

```python
Total sales by product
```

```python
query = """
SELECT 
    PRODUCTCODE, 
    SUM(SALES) AS Total_Sales
FROM sales_data_sample
GROUP BY PRODUCTCODE
ORDER BY Total_Sales DESC;
"""
df = pd.read_sql(query, conn)
df
```

```python
Monthly sales trend
```

```python
query = """
SELECT 
    MONTH_ID, 
    SUM(SALES) AS Monthly_Sales
FROM sales_data_sample
GROUP BY MONTH_ID
ORDER BY MONTH_ID;
"""
df = pd.read_sql(query, conn)
df
```

```python
Top 5 customers by revenue
```

```python
query = """
SELECT TOP 5
    CUSTOMERNAME, 
    SUM(SALES) AS Revenue
FROM sales_data_sample
GROUP BY CUSTOMERNAME
ORDER BY Revenue DESC;
"""
df = pd.read_sql(query, conn)
df
```

```python
Orders by status
```

```python
query = """
SELECT 
    STATUS, 
    COUNT(*) AS Order_Count
FROM sales_data_sample
GROUP BY STATUS;
"""
df = pd.read_sql(query, conn)
df
```

```python
Sales by country
```

```python
query = """
SELECT 
    COUNTRY, 
    SUM(SALES) AS Total_Sales
FROM sales_data_sample
GROUP BY COUNTRY
ORDER BY Total_Sales DESC;
"""
df = pd.read_sql(query, conn)
df
```

```python
 Sales by year and quarter
```

```python
query = """
SELECT 
    YEAR_ID, 
    QTR_ID, 
    SUM(SALES) AS Total_Sales
FROM sales_data_sample
GROUP BY YEAR_ID, QTR_ID
ORDER BY YEAR_ID, QTR_ID;
"""
df = pd.read_sql(query, conn)
df
```

```python
Best-selling products (Top 10 by quantity)
```

```python
query = """
SELECT TOP 10 
    PRODUCTCODE, 
    SUM(QUANTITYORDERED) AS Total_Quantity
FROM sales_data_sample
GROUP BY PRODUCTCODE
ORDER BY Total_Quantity DESC;
"""
df = pd.read_sql(query, conn)
df
```

```python
Average deal size by customer
```

```python
query = """
SELECT 
    CUSTOMERNAME,
    AVG(SALES) AS Avg_Deal
FROM sales_data_sample
GROUP BY CUSTOMERNAME
ORDER BY Avg_Deal DESC;
"""
df = pd.read_sql(query, conn)
df
```

```python
Count of orders by deal size category
```

```python
query = """
SELECT 
    DEALSIZE, 
    COUNT(*) AS Order_Count
FROM sales_data_sample
GROUP BY DEALSIZE
ORDER BY Order_Count DESC;
"""
df = pd.read_sql(query, conn)
df
```

```python
Products with high MSRP but low sales
```

```python
query = """
SELECT 
    PRODUCTCODE, 
    MSRP, 
    SUM(SALES) AS Total_Sales
FROM sales_data_sample
GROUP BY PRODUCTCODE, MSRP
HAVING SUM(SALES) < 10000
ORDER BY MSRP DESC;
"""
df = pd.read_sql(query, conn)
df
```

```python
Customers with more than 10 orders
```

```python
query = """
SELECT 
    CUSTOMERNAME, 
    COUNT(*) AS Total_Orders
FROM sales_data_sample
GROUP BY CUSTOMERNAME
HAVING COUNT(*) > 10
ORDER BY Total_Orders DESC;
"""
df = pd.read_sql(query, conn)
df
```

```python
City-wise revenue distribution
```

```python
query = """
SELECT 
    CITY, 
    SUM(SALES) AS Total_Revenue
FROM sales_data_sample
GROUP BY CITY
ORDER BY Total_Revenue DESC;
"""
df = pd.read_sql(query, conn)
df
```
