```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load the dataset
file_path = r"C:\Users\LENOVO\Desktop\Elevate Labs Internship tasks\Final Project\employee_attrition_dataset_10000.csv"
df = pd.read_csv(file_path)
```

```python

print("Dataset Shape:", df.shape)
df.head()
```

```python
# Import visualization libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Convert 'Attrition' column to binary (Yes=1, No=0)
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
# Group by Department and calculate mean attrition rate
dept_attrition = df.groupby('Department')['Attrition_Flag'].mean().sort_values(ascending=False)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=dept_attrition.values, y=dept_attrition.index, palette="coolwarm")
plt.title("Department-wise Attrition Rate", fontsize=16)
plt.xlabel("Attrition Rate")
plt.ylabel("Department")
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
# Create salary bands using quantile-based binning
df['Salary_Band'] = pd.qcut(df['Monthly_Income'], q=5)
```

```python
# Calculate attrition rate per salary band
salary_attrition = df.groupby('Salary_Band')['Attrition_Flag'].mean()
```

```python
# Plot
plt.figure(figsize=(10, 6))
salary_attrition.plot(kind='bar', color='orange', edgecolor='black')
plt.title("Salary Band-wise Attrition Rate", fontsize=16)
plt.xlabel("Salary Band (Monthly Income)")
plt.ylabel("Attrition Rate")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Group and calculate attrition rate by 'Years_Since_Last_Promotion'
promo_attrition = df.groupby('Years_Since_Last_Promotion')['Attrition_Flag'].mean()
```

```python

plt.figure(figsize=(10, 6))
promo_attrition.plot(kind='line', marker='o', color='green')
plt.title("Attrition vs Years Since Last Promotion", fontsize=16)
plt.xlabel("Years Since Last Promotion")
plt.ylabel("Attrition Rate")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
```

```python
df = pd.read_csv("C:/Users/LENOVO/Desktop/Elevate Labs Internship tasks/Final Project/employee_attrition_dataset_10000.csv")
```

```python
# Convert Attrition to binary
```

```python
df["Attrition_Flag"] = df["Attrition"].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
# --- 1. Department-wise Attrition ---
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Department', y='Attrition_Flag', estimator='mean', palette='Reds')
plt.title("Attrition Rate by Department")
plt.ylabel("Attrition Rate")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print(df.columns.tolist())
```

```python
# --- 2. Salary Band-wise Attrition ---
# Create salary bands
```

```python
df["Salary_Band"] = pd.qcut(df["Monthly_Income"], q=4, labels=["Low", "Medium", "High", "Very High"])
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="Salary_Band", y="Attrition_Flag", estimator='mean', palette='Blues')
plt.title("Attrition Rate by Salary Band")
plt.ylabel("Attrition Rate")
plt.show()
```

```python
# --- 3. Years Since Last Promotion ---
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="Years_Since_Last_Promotion", y="Attrition_Flag")
plt.title("Attrition Rate vs. Years Since Last Promotion")
plt.ylabel("Attrition Rate")
plt.show()
```

```python
# --- 4. Age Group Attrition ---
```
